```python
import requests
import pandas as pd
```

## Comparing Game of Thrones and Star Wars character APIs

### Game of Thrones:

```python
got = requests.get('https://thronesapi.com/api/v2/Characters')
got.status_code
```

```python
got = got.json()
```

```python
got_df = pd.DataFrame(got)
print(got_df.shape)
got_df
```

### Star Wars:

```python
sw = requests.get('https://swapi.dev/api/people')
sw.status_code
```

```python
sw = sw.json()
```

```python
sw_results = sw['results']
```

```python
sw_df = pd.DataFrame(sw_results)
print(sw_df.shape)
sw_df
```

They are both very similar, although the GoT one concentrates on houses and titles, while the Star Wars one gives more of a physical description of the character.
